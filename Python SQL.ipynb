```python
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
```

```python
df.rename(columns = {'City': 'city', 'Order Id': 'order_id'}) # renaming columns like this is tideous task 
```

```python
df.columns = df.columns.str.lower() # hence it should be renamed like this
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.head(5)
```

```python
df['discount'] = df['list_price']*df['discount_percent']*.01
df.head(5)
```

```python
df['sale_price'] = df['list_price']-df['discount']
df['sale_price']
```

```python
df.head(5)
```

```python
df['profit'] = df['sale_price'] - df['cost_price']
df.head(5)
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format = '%Y-%m-%d') # initially order_date dtype was object we changed it to datetime
```

```python
df.drop(columns = ['cost_price', 'discount_percent', 'list_price'], inplace = True)
```

```python
import mysql.connector
import numpy as np

# Connect to MySQL
conn = mysql.connector.connect(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="Deepamshah@7973",
    database="shop"
)
cursor = conn.cursor()

# Create table if not exists
cursor.execute("""
CREATE TABLE IF NOT EXISTS orders (
    order_id INT,
    order_date DATE,
    ship_mode VARCHAR(50),
    segment VARCHAR(50),
    country VARCHAR(100),
    city VARCHAR(100),
    state VARCHAR(100),
    postal_code VARCHAR(20),
    region VARCHAR(50),
    category VARCHAR(50),
    sub_category VARCHAR(50),
    product_id VARCHAR(100),
    quantity INT,
    discount FLOAT,
    sale_price FLOAT,
    profit FLOAT
)
""")

# Replace NaNs with None
df = df.replace({np.nan: None})

# Prepare data and insert
insert_query = """
INSERT INTO orders (
    order_id, order_date, ship_mode, segment, country, city, state, postal_code,
    region, category, sub_category, product_id, quantity, discount, sale_price, profit
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
data = [tuple(row) for row in df.itertuples(index=False)]
cursor.executemany(insert_query, data)

# Finalize
conn.commit()
cursor.close()
conn.close()

print("All records inserted successfully!")
```
